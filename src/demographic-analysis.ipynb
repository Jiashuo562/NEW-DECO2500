Below we begin analysis of the demographic data. It showcases the distribution and frequencies of the following:
1. Genders
2. Occupation
3. Age-Group
4. Usage-Duration

Remark: It is possible to rewrite the following code blocks concisely in one for loop. However, this will
not be done as it will only render the last renderable code block

```python
# Gender Distribution & Frequency
gender_counts = df["gender"].value_counts()
gender_proportion = df["gender"].value_counts(normalize=True) * 100

# Distribution
bar_chart(
    categories=gender_counts.index,
    category_freq= list(gender_counts.values),
    title="Gender Distribution"
) 

# Summary/Frequency
create_freq_table("Gender Frequency Table", gender_counts, gender_proportion)
```

```python
# Occupation Distribution & Frequency
occupation_count = df["occupation"].value_counts()
occupation_prop = df["occupation"].value_counts(normalize=True) * 100

# Distribution
bar_chart(
    categories=occupation_count.index,
    category_freq= list(occupation_count.values),
    title="Occupation Distribution"
) 

# Summary/Frequency
create_freq_table("Occupation Frequency Table", occupation_count, occupation_prop)
```

```python
# Age-Group Distribution & Frequency
age_group_order = ["Under 18", "18-24", "24-34"]

# Use an explicit re-order as the data is ordinal
age_group_counts = df["age_group"].value_counts().reindex(age_group_order)
age_group_props = (age_group_counts / age_group_counts.sum()) * 100 

# Distribution
bar_chart(
    categories=age_group_counts.index,
    category_freq= list(age_group_counts.values),
    title="Age Group Distribution"
) 

# Summary/Frequency
create_freq_table("Age Group Frequency Table", age_group_counts, age_group_props)
```

```python
# Usage-Duration Distribution & Frequency
usage_duration_order = ["Less than 6 months", "6-12 months", "1-3 years", "More than 3 years", "I do not currently use them"]

# We use a custom order again as the data is ordinal
usage_duration_counts = df["usage_duration"].value_counts().reindex(usage_duration_order)
usage_duration_props = (usage_duration_counts / usage_duration_counts.sum()) * 100 

# Distribution
bar_chart(
    categories=usage_duration_counts.index,
    category_freq= list(usage_duration_counts.values),
    title="Usage Duration Distribution"
) 

# Summary/Frequency
create_freq_table("Usage Duration Frequency Table", usage_duration_counts,usage_duration_props)
```

## Demographic Analysis

From the above we arrive at the following:
1. Surveyees were not primarily from one gender group and was roughly equal
2. A significant amount of surveyees were university students, which was naturally expected as we mainly asked around for surveyees in school
3. Unsurprisingly the largest age-group surveyed coincides with the expected age-group of current university students
4. About half of the surveyed no longer use wearables. However, their willigness to participate in the survey hints to possible unmet needs that caused them to no longer use wearables
